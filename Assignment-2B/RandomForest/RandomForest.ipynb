# Random Forest Traffic Volume Prediction

This notebook trains a global long-format Random Forest model so the third model has a stronger classical-ML baseline than a single decision tree.

```python
from RandomForest_model import (
    get_locations,
    save_random_forest_plots,
    traffic_random_forest,
    train_random_forest_model,
)
```

```python
locations = get_locations()
print("Locations found:", locations)
```

```python
results = train_random_forest_model(eval_loc_index=1, eval_time_index=0)
print(f"Global RMSE:  {results['rmse']:.2f}")
print(f"Global NRMSE: {results['nrmse']:.2f}")
if results['eval_rmse'] is not None:
    print(f"Focused location: {results['eval_location']}")
    print(f"Focused time slot: V{results['eval_time_index']:02d}")
    print(f"Focused RMSE:  {results['eval_rmse']:.2f}")
    print(f"Focused NRMSE: {results['eval_nrmse']:.2f}")
```

```python
plot_paths = save_random_forest_plots(results)
print(plot_paths)
```
